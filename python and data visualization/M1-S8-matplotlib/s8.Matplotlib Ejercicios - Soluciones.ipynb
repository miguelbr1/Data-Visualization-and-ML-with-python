# Ejercicios

Siga las instrucciones para recrear las siguientes figuras:

## Datos

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

**Importe matplotlib.pyplot como plt y configure "%matplotlib inline" para los notebooks de jupyter.**

```python
import matplotlib.pyplot as plt
%matplotlib inline
# plt.show() # no-notebook
```

## Ejercicio 1

**Siga estos pasos:**
* **Cree un objeto figura llamado fig usando plt.figure()**
* **Use add_axes para agregar un eje al canvas de la figura en [0,0,1,1]. Llame a este nuevo eje ax.**
* **Grafique "plot(x, y)" en ese ejes y configure las etiquetas y títulos para que coincidan con la figura que se presenta a continuación:**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

## Ejercicio 2

**Crear un objeto figura y poner dos ejes sobre él, ax1 y ax2. Ubicados en [0,0,1,1] y [0.2,0.5,.2,.2], respectivamente.**

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
```

**Ahora grafique (x, y) en ambos ejes. Configure las etiquetas de los ejes y llame al objeto figura para mostrarlo.**

```python
ax1.plot(x,y)
ax1.set_xlabel('x')
ax1.set_ylabel('y')


ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')

fig # Muestra la figura
```

## Ejercicio 3

**Crear la gráfica que se presenta a continuación agregando dos ejes a un objeto de figura en [0,0,1,1] y [0.2,0.5,.4,.4]**

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
```

**Ahora use los arreglos 'x', 'y' y 'z' para recrear la gráfica siguiente. Observe los límites de los ejes en el gráfico interior:**

```python
ax.plot(x,z)
ax.set_xlabel('X')
ax.set_ylabel('Z')


ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)

fig
```

## Ejercicio 4

**Use plt.subplots (nrows=1, ncols=2) para crear el diagrama que se muestra a continuación.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

**Ahora grafique (x,y) y (x,z) en esos ejes. Configure adecuadamente los parámétros color, linewidth y linestyle para obtener lo siguiente:**

```python
axes[0].plot(x,y,color="blue", lw=3, ls='--')
axes[1].plot(x,z,color="red", lw=3, ls='-')
fig
```

**Redimensione la figura incluyendo el parámetro figsize al momento de crear la figura en plt.subplots(). Copie y pege el código de la celda anterior para ahorrar tiempo.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12,2))

axes[0].plot(x,y,color="blue", lw=5)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x,z,color="red", lw=3, ls='--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```
