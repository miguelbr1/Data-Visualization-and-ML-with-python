```
# Importamos librerías
import numpy as np
from bokeh.plotting import figure 
from bokeh.io import output_notebook, show, output_file, save, reset_output
output_notebook()
```

## 1. Gráfico de dispersión

Haga un gráfico de dispersión con la función sigmoide, para los valores de x entre [-100, 100].

Definida por la siguiente fórmula: 

$y=\frac {1}{1+e^{-x}}$

Su gráfica debe verse así: 

![texto alternativo](https://drive.google.com/uc?export=download&id=1OD-G06Ru5x3tO1wIE-2JHpJ5oa0-4MJg)

```
# ESCRIBA SU CODIGO AQUI
# Solución

x = np.linspace(-10, 10, 100)
sig = lambda x : 1/(1+np.exp(-x))
y = sig(x)
p = figure(width=500, height=500)
p.circle(x, y, size=10, color="firebrick", alpha=0.5)
#output_file("ejemplo.html")
output_notebook()
show(p)
```

## 2. Gráfico de dispersión con variable categórica


Haga un gráfico de dispersión con el dataset de Iris anteriormente cargado en dónde el eje x corresponda con el largo del pétalo y el eje y con el ancho del pétalo.

Su gráfica debe verse así:

![texto alternativo](https://drive.google.com/uc?export=download&id=1wYCQrU2fgmiSxg_ibMzwqz2vd4ly2gGO)

```
# Cargamos los datos
from bokeh.sampledata.iris import flowers as df
df.head()
```

```
# ESCRIBA SU CÓDIGO AQUÍ
#solución

colormap = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
df['color'] = df['species'].map(lambda x: colormap[x])

output_file("iris.html", title="iris.py example")

p = figure(width=500, height=500)
p.xaxis.axis_label = 'Petal Length'
p.yaxis.axis_label = 'Petal Width'

p.circle(df["petal_length"], df["petal_width"],
        color=df["color"], fill_alpha=0.2, size=10, )

output_file("ejemplo.html")
output_notebook()
show(p)
```
