** ¿Cuánta gente tiene la palabra 'Chief' en su cargo?**

```python
"Chief".lower() in "MasterChief".lower()
```

```python
# ESCRIBA SU CODIGO AQUI PARA NO BORRAR EL RESULTADO ESPERADO
def findChief(word):
  word = word.lower()
  return "Chief".lower() in word
```

```python
findChief("MasterChief")
```

```python
# ESCRIBA SU CODIGO AQUI PARA NO BORRAR EL RESULTADO ESPERADO
sal["JobTitle"].apply(findChief).sum()
```
